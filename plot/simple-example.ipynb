```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
plt.show()
```

```python
fig.savefig('simple-graph.png')
```
